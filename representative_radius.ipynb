What is the representative length of a collection of volumes with a uniform distribution of lengths between $x_{low}$ and $x_{high}$?

Disregarding whether the volumes are spheres, cubes, or some other shape, the relationship between length and volume is cubic:

$$
V = k \cdot x^3
$$



```python
import numpy as np 
import matplotlib.pyplot as plt
import random
import math
%matplotlib widget
```

```python
def v_avg(x_low, x_high):
    return math.pi * (x_high**2 + x_low**2) * (x_high + x_low) / 3

# def p(x_low, x_high):
#     return (v_avg(x_low, x_high)**(1/3) - x_low) / (x_high - x_low)

# def p(x_low, x_high, epsilon=1e-10):
#     q = v_avg(x_low, x_high) ** (1 / 3)
#     n = q * (x_high + x_low) - x_low**2 - x_low * x_high
#     d = x_high**2 - x_low**2

#     return n / d

def p(x_low, x_high):
    q = (3* v_avg(x_low, x_high) / 4 / math.pi )** (1 / 3)
    return q/x_high

# def x_rep(x_low, x_high):
#     x_ratio = x_high / x_low
#     return (x_high**3/4 * (1 + x_ratio**2) * (1 + x_ratio))**(1/3)



def x_rep(x_low, x_high):
    a = x_high**2 + x_low**2
    b = x_high + x_low
    return (1 / 4 * a * b) ** (1 / 3)
```

```python
epsilon = 1e-10
x_low = 0 #+ epsilon
x_high = 1
x = np.linspace(x_low, x_high, 11)
v = 4*math.pi/3 * x**3
```

```python
plt.close("all")
fig, ax = plt.subplots()
ax.scatter(x, v)
# ax.plot(x, v_avg_1(x, v), label='Average Volume')
ax.plot(x, v_avg(x, x_high)/max(v), label="Average Volume (Method 2)")
ax.plot(x, p(x, x_high), label = "p")
ax.plot(x, x_rep(x, x_high), label="x rep")
# ax.plot(x, v_avg_3(x, x_high), label="Average Volume (Method 3)")
ax.set_xlim(0, x_high + 1)
# ax.set_ylim(0, x_high**3 + 1)
# ax.set_aspect('equal', adjustable='box')
ax.set_xlabel("Length")
ax.set_ylabel("Volume")
ax.legend()
ax.grid(True)
plt.show()
```
